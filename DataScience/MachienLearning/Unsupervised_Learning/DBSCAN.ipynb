```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dfraw = pd.read_csv('/home/sam/Desktop/Codes/DataScience/datasets/Mall_Customers.csv')
df = dfraw.rename({'Annual Income (k$)' : 'income', 'Spending Score (1-100)' : 'score'}, axis=1)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=5, min_samples = 5)
```

```python
prediction = model.fit_predict(df[['income', 'Age']])
df['label'] = prediction
```

```python
# visualization
a = df['label'] == 1
plt.figure(figsize = (15, 8))
sns.color_palette('viridis', as_cmap= True)
sns.scatterplot(df['income'], df['score'], s=50, hue=df['label'])
```
