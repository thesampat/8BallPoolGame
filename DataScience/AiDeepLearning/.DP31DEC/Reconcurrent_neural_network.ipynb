```python
import numpy as np
import copy
np.random.seed(0)

x1 = np.array([1, 1, 0, 1])
x2 = np.array([0, 1, 0, 1])
y1 = np.array([1, 1, 0, 1])

# model features
in_layer = 2
hidden_layer = 4
out_layer = 1

# activations
def sigmoid(x):
    output=  1/(1+np.exp(-x))
    return output
    

def der_sigmoid(x):
    return x*(1-x)

# weights and update prams
wx = 2*np.random.random((in_layer, hidden_layer)) -1
wy = 2*np.random.random((hidden_layer, out_layer)) - 1
wh = 2*np.random.random((hidden_layer, hidden_layer)) -1

wx_update = np.zeros_like(wx)
wy_update = np.zeros_like(wy)
wh_update = np.zeros_like(wh)

# network parmameters
lr = 0.1
epochs = 500

# model
for epoch in range(epochs):
    hidden = list()
    deltal = list()
    pred = list()
    h = np.zeros(hidden_layer)
    hidden.append(h)
    
    for i in range(x1.shape[0]):
        X = np.array([x1[i], x2[i]])
        Y = np.array(y1[i])
        k = np.dot(h, wh)
        h = sigmoid(np.dot(X, wx) + np.dot(h, wh))
        a = sigmoid(np.dot(h, wy))
        
        loss = Y -a
        delta = loss * der_sigmoid(a)
        deltal.append(delta)
        
        hidden.append(h)
        future_layer = np.zeros(hidden_layer)
        pred.append(a)
        
    print(pred)
    for j in range(x1.shape[0]-1, -1, -1):
        current_neuron = np.array(hidden[j+1]).reshape(1, 4)
        prev_neuron = np.array(hidden[j]).reshape(1, 4)
        X = np.array([x1[j], x2[j]]).reshape(2, 1)
        
        n_delta = deltal[j].reshape(1,1)
        d_e_out = (future_layer.dot(wh.T) + n_delta.dot(wy.T)) * der_sigmoid(current_neuron)
        wy_update += current_neuron.T.dot(n_delta)
        wh_update += np.atleast_2d(prev_neuron).T.dot(d_e_out)
        wx_update += X.dot(d_e_out)
        

    wx += wx_update * lr
    wy += wy_update * lr
    wh += wh_update * lr

    wx_update *= 0
    wy_update *= 0
    wh_update *= 0
```
