```python
import copy
import numpy as np
np.random.seed(0)

# data
x1 = np.array([1, 0]).reshape(2, 1)
x2 = np.array([1, 1]).reshape(2, 1)
y1 = np.array([0, 1]).reshape(2, 1)

# properties
layerx = 2
layerh = 6
layero = 1

# weights 
# initialize neural network weights
wx = 2 * np.random.rand(layerx, layerh) -1
wy = 2 * np.random.rand(layerh, layero) -1
wh = 2 * np.random.rand(layerh, layerh) -1

wx_update  = np.zeros_like(wx)
wy_update  = np.zeros_like(wy)
wh_update  = np.zeros_like(wh)

def sigmoid(x):
    output = 1/(1+np.exp(-x))
    return output

def der_sigmoid(x):
    return x*(1-x)

def forward(x, h):
    h = sigmoid(np.dot(x.T, wx) + np.dot(h, wh))
    y = sigmoid(np.dot(h, wy))
    return h, y

epochs = 500
lossi = np.zeros_like(y1)
lr = 0.5

for epoch in range(epochs):    
    hidden = list()
    layer2_delta = list()
    h = np.zeros(layerh)
    hidden.append(np.zeros(layerh))
    pred = list()
    for i in range(len(x1)):
        x = np.array((x1[i], x2[i]))
        y = y1[i]
        h, pred_y = forward(x, h)
        loss = (y - pred_y)
        pred.append(pred_y)
        delta = loss * der_sigmoid(pred_y)
        hidden.append(copy.deepcopy(h))
        
        layer2_delta.append(delta)
        feature_delta  = np.zeros(layerh)
    print(pred)

        
    for j in range(len(x)-1, -1, -1):
#         X = np.array([x1[j],x2[j]])
        current_hidden  =  hidden[j+1]
        previous_hidden = hidden[j]
        layer_2_delta = layer2_delta[j]
        current_neuron = feature_delta.dot(wh.T) + layer_2_delta.dot(wy.T) * der_sigmoid(current_hidden)
        wy_update += np.atleast_2d(current_hidden).T.dot(layer_2_delta)
#         wh_update += np.atleast_2d(previous_hidden).T.dot(current_neuron)
#         wx_update += X.dot(current_neuron)

#         feature_delta = current_neuron
#         feature_hidden = current_hidden
        
    wx += wx_update * lr
    wy += wy_update * lr
    wh += wh_update * lr

    wx_update *= 0
    wy_update *= 0
    wh_update *= 0
```
